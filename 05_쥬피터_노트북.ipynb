# 쥬피터 노트북(Jupyter Notebook)

쥬피터 노트북은 프로그램 코드와 실행 결과, 수식, 그래프를 포함하는 문서를 작성하고 공유할 수 있게 해주는 웹 기반의 응용프로그램이다.

Note: Jupyter는 Julia, Python, R을 조합한 약자이다.

연구자들은 보통 논문이나 보고서 파일, 자료분석 프로그램 파일, 프로그램 실행 결과 파일을 별도로 관리한다.
일반적으로 논문이나 보고서에는 자료와 자료분석 프로그램이 거의 포함하지 않다.
따라서 논문이나 보고서를 보는 사람은 자료분석 과정에 대해 전혀 알지 못하고 분석 결과를 확인할 길이 거의 없다.
이런 문제를 해결하기 위해 등장한 개념이 **reproducible research**라는 개념이며 쥬피터 노트북은 이를 가능하게 해준다.

그러면 웹 브라우저에서 폴더, 텍스트 파일, 쥬피터 노트북을 생성/수정/삭제할 수 있고 쥬피터 노트북 문서 내에서 프로그램을 작성/실행할 수 있다.

## 쥬피터 노트북 사용법

1. 대쉬보드에서 `New` 버턴을 누르고 `Python 3`을 선택하여 새로운 쥬피터 노트북을 생성한다.
(다른 언어로 작성한 코드를 실행할 수 있는 커널을 설치하면 다른 언어의 코드를 실행하는 쥬피터 노트북을 생성할 수 있다.)
**Google Drive에서 오른쪽 마우스 버턴을 누르고, 메뉴에서 Colaboratory를 선택하면 새로운 Jupyter Notebook이 생성된다.**

2. 생성된 쥬피터 노트북은 삭제, 이름 변경, 복사, 다른 형식으로 export할 수 있다.
3. 쥬피터 노트북은 셀(cell)로 구성되며, 두 종류의 셀을 주로 사용한다. 프로그래밍을 위한 코드 셀과 문서 작성을 텍스트(text) 셀이 그것이다.
코드 셀의 타입은 `code`이고 텍스트 셀의 타입은 `markdown`이다.

4. `markdown`은 텍스트 셀가 에서 문서를 작성할 때 사용하는 언어를 가르키는 용어이다. 그러니까 텍스트 셀에는 `markdown`을 이용하여 문서를 작성한다. 하지만 `markdown`이 수식은 지원하지 않기 때문에  수식은 `LaTeX` 양식으로 입력한다.

5. Python 프로그램은 코드 셀에 입력한다.

6. 코드 셀에 입력한  프로그램을 실행 또는 재실행할 때는  해당 셀을 선택한 다음 `ctrl+enter`를 누른다.

7. 셀을 삭제, 수정하거나 문서 내에서 이동할 수 있고, 셀의 타입도 변경할 수 있으며, 셀을 분할하거나 합병할 수도 있다.

## 마크다운

웹의 등장으로 HTML, XML과 같은 ``markup language``가 등장하였지만 이들은 익히고 사용하는데 상당한 시간이 걸리며 그 문서를 읽기 어렵다.
이들과 달리 작성하고 읽기 쉬운 ``markup language``로 등장한 것이 **마크다운**(markdown)이다.

마크다운을 이용하면 메모장과 같은 간단한 편집기로도 질 좋은 문서를 만들 수 있으며, 적절한 프로그램을 사용하면 HTML, LaTeX 등으로 변환할 수도 있다.

쥬피터 노트북의 텍스트 셀의 타입이 **markdown**인 이유는 문서를 마크다운으로 작성하기 때문이다.
하지만 필요하면 HTML tag를 사용해도 된다.

## 마크다운으로 문서 작성하는 방법

1. 제목줄: # 기호로 시작한다. #의 개수가 많아지면 글자 크기가 작아진다.

2. 한 줄 이상의 빈 줄이 있으면 문단이 바뀐다.

3. 밑줄 또는 별표로 감싼 부분은 *이탤릭체*가 된다.

4. 두 개의 별표로 감싼 부분은 **굵은 글씨체**가 된다.

5. 목록을 만들 때는 목록의 항목을 **별표**로 시작하면 된다.
목록의 각 항목을 들여쓰기 하면 중첩된 목록을 만들 수 있다.
    * 항목 1
    * 항목 2
    
6. 목록 항목에 일련번호를 붙이고 싶으면 목록의 항목을 별표 대신 **정수**로 시작하면 된다.
이때 정수를 반드시 순서대로 줄 필요가 없다.
    3. 항목 1
    4. 항목 2
    
7. **url link**를 문서에 포함시키고 싶으면 `[링크이름](url)`라고 하면 된다. 마크다운에 대한 자세한 설명을 보고 싶으면 [여기](http://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Working%20With%20Markdown%20Cells.html)나 [여기](https://daringfireball.net/projects/markdown/)를 클릭하시오.

8. 문서에 이미지를 포함시키고 싶으면 `![이미지에대한텍스트](경로를포함한이미지파일이름)`와 같이 하면 된다.
![파이썬 로고](./images/python-logo.png)

10. 수식은 LaTeX의 문법에 따라 입력한다.
본문의 수식은 \$로 감싸고 독립적인 수식은 \$\$로 감싼다.
예를 들어 $x \in R$에 대해서 다음 함수를 생각하자.
$$
f(x) = ax^2 + bx + c
$$
LaTeX에 관한 도움말은 [여기](http://www.andy-roberts.net/writing/latex)를 참고하기 바란다.

9. 표를 작성할 때 열의 시작과 끝은 파이프(`|`)로 표시한다.
표의 첫 줄에는 열 이름을 기재하고, 다음 줄에는 각 열의 정렬을 지정한다.
각 열의 폭은 자동적으로 결정된다.
열의 시작 파이프 다음에 콜론(`:`)을 두면 이 열은 좌측 정렬, 열의 끝을 알리는 파이프 앞에 콜론을 쓰면 해당 열은 우측 정렬한다.
만약 열의 시작과 끝 파이프에 모두 콜론을 쓰면 가운데 정렬된다.
(정렬은 지원되지 않는 경우가 있다.)

| 이름 | 학과 |
|-----:|:-----|
|박중양|정보통계학과|
| 123 | 678.90|



## 쥬피터 노트북을 다른 형식의 문서로 변환하기

쥬피터 노트북은 HTML, TeX 파일로 변환할 수 있다.
LaTeX 사용자라면 TeX 파일로 변환하여 질 좋은 pdf 파일로 변환할 수 있다.

### HTML 파일로 변환하기

명령창을 실행하고 쥬피터 노트북이 저장된 폴더로 이동한 다음 아래 명령을 실행하면 연습문제에서 만든 쥬피터 노트북 파일을 HTML 파일로 변환할 수 있다.

```python
C:\Users\joongyang> jupyter nbconvert --to html '쥬피터노트북.ipynb'
```

### 발표자료로 변환하기

HTML 파일로 변환할 때와 마찬가지로 `nbconvert`를 이용하면 발표용 HTML 파일로 변환할 수 있다.

```python
C:\Users\joongyang> jupyter nbconvert --to slides '쥬피터노트북.ipynb'
```

`nbpresent` 도구를 이용해서 발표용 HTML 파일로 변환할 수도 있다.
먼저 `nbpresent`를 설치해야 하는데 다음과 같이 Anaconda의 `conda` 명령으로 설치하면 된다.

```python
conda install -c conda-forge nbpresent
```

`nbpresent` 도구를 이용해서 쥬피터 노트북을 발표용 HTML 파일로 변환할 때는 명령창에서 다음 명령을 실행한다.

```python
C:\Users\joongyang> nbpresent -i 쥬피터노트북.ipynb -o 발표자료.html
```

### MS-Word 파일로 변환하기

[pandoc](https://pandoc.org/installing.html)을 이용하면 HTML 파일을 확장자가 docx인 MS-Word 파일로 변환할 수 있다.

```python
C:\Users\joongyang> pandoc -f html -t docx -o 쥬피터노트북.docx '쥬피터노트북.ipynb'
```
