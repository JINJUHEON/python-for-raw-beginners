# `List Comprehension`

파이썬에는 `list comprehension`이라는 기존하는 자료구조에서 새로운 자료구조를 생성하는 방법이 있다.
자료구조에 저장된 각 값을 이용해서 구한 새로운 값을 저장한 새로운 자료구조를 생성할 때 사용하는 방법이다.
기본적인 사용법은 다음과 같으며 자료구조 data_struct에 저장된 각 값을 ele로 받아서 표현식을 계산한 결과를 저장한 새로운 리스트가 생성된다.

```python
new_data_struct = [ ele를포함하는표현식 for ele in data_struct ]
```

`list comprehension`은 data_struct가 리스트가 아닌 자료구조인 경우에도 사용할 수 있으며, 새로 생성되는 자료구조도 `[]` 대신 어떤 괄호를 쓰느냐에 따라 달라진다.
위처럼 `[]`를 사용하면 결과는 리스트가 된다.

$0 \le x \le 10$ 구간을 100등분하는 101개 값에서 함수 $f \left( x \right) = x^2 + 2x + 3$의 값은 다음과 같이 구할 수 있다.

````python
import math
xx = list(range(0, 101))

x = []
for xx0 in xx:
    x.append(xx0 / 10)

a = 1
b = 2
c = 3
fx = []
for x0 in x:
    fx.append(a * math.pow(x0, 2) + b * x0 + c)
print(fx)
````

`list comprehension`을 이용하면 다음과 같이 간단히 프로그램 할 수 이다.

````python
x = list(range(0, 101))

x = [ x0 / 10 for x0 in x ]

a = 1
b = 2
c = 3

fx = [ a * math.pow(x0, 2) + b * x0 + c for x0 in x ]

print(fx)
````

다음은 `list comprehenison`과 `if` 문장을 결합하는 방법이다.

```python
new_data_struct = [ ele를포함한표현식 for ele in data_struct if ele를포함한조건 ]
```

이 방법은 자료구조에 저장된 값 중에서 특정한 조건을 만족하는 값만 선택해서 새로운 값을 계산하고 그 결과를 담은 자료구조를 생성할 때 사용한다.
예를 들어, $0 \le x \le 10$ 구간을 100등분하는 101개 값에서 구한 함수 $f \left( x \right) = x^2 + 2x + 3$의 값 중에서 50과 100 사이의 값만 찾아서 새로운 리스트를 만들 수 있다.

````python
fx_btw_50_100 = [ fx0 for fx0 in fx  if 50 <= fx0 <= 100 ]
print(fx_btw_50_100)
````

만약 `list comprehension`을 사용하지 않으면 다음과 같이 프로그래밍 해야 할 것이다.

````python
fx_btw_50_100 = []
for fx0 in fx:
    if 50 <= fx0 <= 100:
        fx_btw_50_100.append(fx0)
print(fx_btw_50_100)
````

`list comprehension`과 `if` 문장을 결합할 때 다음과 같이 `else` 절도 포함할 수 있다.

```python
new_list = [ele를포함한표현식1 if ele를포함한조건 else ele를포함한표현식2 for ele in data_struct]
```

이 방법은 자료구조의 각 값에 대해 `if` 절의 조건을 평가하고 그 결과가 `True`이면 `ele를포함한표현식1`를 계산하고 `False`이면 `ele를포함한표현식2`를 계산한 결과를 담은 자료구조를 생성한다.
예를 들어, $0 \le x \le 10$ 구간을 100등분하는 101개 값에서 구한 함수 $f \left( x \right) = x^2 + 2x + 3$의 값 중에서 50과 100 사이의 값은 그대로 두고 나머지는 0으로 대체한 리스트는 다음과 같이 만들 수 있다.

````python
new_fx = [ fx0 if 50 <= fx0 <= 100 else 0 for fx0 in fx ]
print(new_fx)
````

만약 `list comprehension`을 사용하지 않았다면 다음과 같이 프로그래밍 해야 할 것이다.

````python
new_fx = []
for fx0 in fx:
    if 50 <= fx0 <= 100:
        new_fx.append(fx0)
    else:
        new_fx.append(0)

print(new_fx)
````

## `list comprehension`을 이용한 자료처리

파이썬의 강력한 기능 중 하나인 `list comprehension`은 가능하면 사용하는 것이 바람직하다.
자료를 처리할 때도 `list comprehension`은 상당히 효율적이다.
**대사율자료**를 이용해서 `list comprehension`의 사용법을 익혀보기로 하자.

````python
gender = ['M']*2 + ['F']*4 + ['M'] + ['F']*2 + ['M'] + \
        ['F']*2 + ['M'] + ['F']*4 + ['M']*2
weight = [62.0, 62.9, 36.1, 54.6, 48.5, 42.0, 47.4, \
          50.6, 42.0, 48.7, 40.3, 33.1, 51.9, 42.4, \
          34.5, 51.1, 41.2, 51.9, 46.9]
metabolic_rate = [1792, 1666, 995, 1425, 1396, 1418, 1362, 1502,\
        1256, 1614, 1189, 913, 1460, 1124, 1052, 1347,\
        1204, 1867, 1439]
````

50 이상이고 60 미만인 체중은 다음과 같이 구할 수 있다.

````python
weight_btw_50_60 = [ wgt for wgt in weight if 50 <= wgt < 60 ]
print(weight_btw_50_60)
````

체중이 50 이상이고 60 미만인 경우의 대사율과 성별은 는 다음과 같이 구할 수 있다.

````python
gender_btw_50_60 = [ gender[idx] for idx, wgt in enumerate(weight) if 50 <= wgt < 60 ]
metabolic_rate_btw_50_60 = [ metabolic_rate[idx] for idx, wgt in enumerate(weight) if 50 <= wgt < 60 ]
print(gender_btw_50_60)
print(metabolic_rate_btw_50_60)
````
