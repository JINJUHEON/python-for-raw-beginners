## 2020년 1학기 강의 소개
---

코로나 바이러스로 인해 이번 학기 강의를 진행하는데 여러 가지 어려움이 생겼습니다.
수강 신청을 한 학생들은 이 강의에 대한 소개와 진행 방법을 참고하여 계속 소강 또는 수강 변경 여부를 신중하게 결정하시기 바랍니다.

1. 이 강의는 제가 강의하는 `정보 통계학 실습`의 연속 과목으로 개설된 것입니다.

2. 따라서 [GitHub에 올린 강의 자료](https://github.com/joongyang/python-for-raw-beginners) 중에서 `13_자료구조_리스트의_이용`까지 이해하고 있는 것으로 간주하고 강의를 할 계획입니다. 그리고 `27_파일처리_연습문제`까지 강의하는 것이 목표입니다.

3. 강의 첫 2주 동안에는 `13_자료구조_리스트의_이용`까지의 내용을 복습할 계획이므로 제가 강의하는 `정보 통계학 실습`을 수강하지 않은 학생일지라도 첫 2주 동안  `04_파이썬_설치와_실행`에서 `13_자료구조_리스트의_이용`까지의 집중적으로 공부한다면 수강할 수 있을 것으로 생각합니다. `No pain, no gain`입니다.

4. 수업은 강의와 실습을 동시에 하는 방식이기 때문에 컴퓨터가 설치된 실습실에서 진행됩니다. 학기가 진행됨에 따라 고장이 나는 컴퓨터가 생기기 때문에 수강생들은 가능하면 자신의 컴퓨터를 가지고 올 것을 권장합니다.

5. 파이썬을 설치하고 관리하는데 시간을 소모하지 않기 위해 Google에서 제공하는 클라우드 서비스를 이용하여 실습을 할 것입니다. 따라서 수강생들은 구글 계정이 있어야 하며 구글 드라이브를 사용할 줄 알아야 합니다. 이에 대한 강의는 없습니다.

6. 시험은 온라인으로 실시합니다. 시험 문제는 쥬피터 노트북으로 제공하며, 수강생들은 답안을 추가한 쥬피터 노트북을 메일에 첨부하여 보내는 방법입니다.
평소 3번의 온라인 시험을 실시했지만 이번 학기에는 2번 보는 것으로 결정하였습니다. 각 시험은 반영 비율은 40%이며, 출석 점수의 반영 이율은 10%, 과제의 반영 비율은 10%입니다.

7. 이번 학기 첫 2주는 비대면 강의이므로 `04_파이썬_설치와_실행`에서 `13_자료구조_리스트의_이용`까지의 강의 자료를 읽고 각 강의 자료에 있는 연습문제를 풀어서 제출해야 합니다. 다음 일정에 따라서 강의 자료로 공부한 다음 연습문제를 푼 쥬피터 노트북을 해당 날짜에 메일에 첨부하여 제출하여야 출석을 인정합니다. 보낼 메일 주소는 `joongyang.park@gmail.com`이며, 메일의 제목은 반드시 `통계프로그래밍 학과 학번 이름`으로 하여야 합니다.
